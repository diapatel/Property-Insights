### society

```python
houses['society'].value_counts()
# 185 unique values
```

```python
houses['society'] = houses['society'].str.lower()
```

```python
houses['society'].value_counts()
```

```python
houses['society'] = houses['society'].str.strip().str.replace(r'\d+\.\d+\s★', '', regex=True)
```

```python
houses['society'].isnull().sum()
```

```python
# replacing null values with "independent"
houses['society'] = houses['society'].fillna('independent')
```

### price

```python
houses['price'].value_counts()
```

```python
houses[houses['price'] == 'Price on Request'].count()
```

- 10 rows where price = "Price on Request". dropping them

```python
houses = houses[houses['price'] != 'Price on Request']
```

```python
houses.shape
```

```python
houses['price']

def handle_price(x):
    if isinstance(x,str):
        if 'Lac' in x:
            return round((float(x.split(' ')[0])/100),2)
        else:
            return round(float(x.split(' ')[0]),2)
```

```python
houses['price'] = houses['price'].apply(handle_price)
```
